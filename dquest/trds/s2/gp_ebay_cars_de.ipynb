#### Clean column names

Define a function `clean_col(str)` to apply various string transformation for each column in the list of columns

define a function `camel_to-snake(text)` to perform format transformation based on regular expressions

```python
def camel_to_snake(text):
    import re
    str1 = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', text)
    return re.sub('([a-z0-9])([A-Z])', r'\1_\2', str1).lower()

def clean_col(col):
    col = col.strip()
    col = col.replace("yearOfRegistration", "registration_year")
    col = col.replace("monthOfRegistration", "registration_month")
    col = col.replace("notRepairedDamage", "unrepaired_damage")
    col = col.replace("dateCreated", "ad_created")
    col = camel_to_snake(col)
    return col

autos.columns = [ clean_col(c) for c in autos.columns ]

pp(autos.columns)
```

```python
autos.drop("nr_of_pictures",axis=1, inplace=True) # this column does not contain useful info
```

```python
# transform text to floats
autos["price"] = autos["price"].str.split('$',expand=True).iloc[:,1].str.replace(',','').astype(float)
autos["odometer"] = autos["odometer"].str.replace("km",'').str.replace(',','').astype(float)
autos.rename(columns={'odometer':'odometer_km'}, inplace=True)

autos[["price","odometer_km"]].describe()

#odometer_km seems to be good,  price has a high outlier
```

```python
autos.price.max()  # we have a high max probably an outlier. 
```

```python
#lets explore the high end of prices
autos[autos.price > 500000].loc[:,["name", "odometer_km","price" ]]
# we have ferraris which seems to be correct, the rest is false data
```

```python
autos[((autos.price > 500000) & ~(autos.name.str.contains("Ferrari")) )]
#dropping thos values except  for ferraris
autos=autos[~((autos.price > 500000) & ~(autos.name.str.contains("Ferrari")) )] #antimasking

autos.shape #stripped off 12 rows
```

```python
#converting  ad_created and last_seen to datetime
autos.loc[:,"ad_created"] = pd.to_datetime(autos.loc[:,"ad_created"])
autos.loc[:,"last_seen"] = pd.to_datetime(autos.loc[:,"last_seen"])
```
