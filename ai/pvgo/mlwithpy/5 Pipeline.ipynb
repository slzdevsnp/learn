```python
%matplotlib inline

import numpy as np
import pandas as pd
from sklearn import datasets
```

```python
from sklearn.pipeline import Pipeline
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
# look for nulls
pd.DataFrame(X).count()
```

## Artificially dirty the data by adding nulls

```python
#  dirty the data by replacing 20% of the values with null (randomly)
np.random.seed(10)
mask = np.random.choice([True, False], size=X.shape, p=[.2,.8])
new_X = pd.DataFrame(X).mask(mask)
```

```python
# verify that the counts are different (there are nulls in teh data)
new_X.count()
```

```python
from sklearn.preprocessing import Imputer

from sklearn.linear_model import LogisticRegression

from sklearn.model_selection import train_test_split
```

```python
imp = Imputer()
```

```python
clf = LogisticRegression()
```

```python
pipeline = Pipeline(  #check  Pipeline object
    [
        ("impute", imp), 
        ("logreg", clf)
    ]
)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    new_X, y, test_size=0.30, random_state=3)
```

```python
pipeline.fit(X_train, y_train)
```

```python
pipeline.score(X_test, y_test)
```

```python
# show input features along side prediction with confidence score

prediction_table = X_test.copy()
prediction_table['real'] = y_test
prediction_table['predict'] = pipeline.predict(X_test)
prediction_table['confidence'] = pipeline.predict_proba(X_test).max(axis=1)
prediction_table['correct'] = np.equal(prediction_table['real'].values, 
                                       prediction_table['predict'].values)
```

```python
prediction_table.sort_values(by='confidence')
```
