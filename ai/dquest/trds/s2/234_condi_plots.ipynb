```python
import pandas as  pd
from pprint import pprint as pp
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

#ignore all warnings
import warnings
warnings.filterwarnings('ignore')
```

## Seaborn with FaceGrids

```python
#titanic data
fn='titanic/train.csv'
tfolks = pd.read_csv(fn)
pp(tfolks.isnull().sum() )
pp(tfolks.shape)

#leave specific cols, and drop rows with NA values
num_cols = ["Survived", "Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]
titanic = tfolks[num_cols]
titanic.dropna(inplace=True) #drop rows with NaS  ?
pp(titanic.shape)
```

### fig 4. simple hist  and pdf

```python
sns.kdeplot(titanic["Age"],shade=True) #just  a pdf 
plt.xlabel("Age")
plt.show()
```

```python
sns.set(style="white")
sns.distplot(titanic["Age"],kde=False)
plt.show()
```

#### fig 5. pdf with aestetics

```python
sns.set_style("white")
sns.kdeplot(titanic["Age"],shade=True)
sns.despine(top=True, right=True, left=True,bottom=True)  #no axe frame
plt.xlabel("Age")
plt.show()
```

#### fig 6. facegride on Pclass categorical variable column

```python
# titanic['Pclass'].value_counts()   is a categorical vaariable with 3 classes
g = sns.FacetGrid(titanic, col="Pclass", size=6)  
sns.despine(top=True, right=True, left=True,bottom=True)
# For each subset of values, generate a kernel density plot of the "Age" columns.
g.map(sns.kdeplot, "Age", shade=True)
#g.map(plt.hist, "Age")  #for histograms
plt.show()
```

#### fig 7.  facegrid on 2 params makes a a 2 x 3 grid

```python
g = sns.FacetGrid(titanic, row="Pclass", col="Survived", size=3)
g.map(sns.kdeplot, "Age", shade=True)
plt.show()
```

### fig  8. 

```python

g = sns.FacetGrid(titanic, row="Pclass", col="Survived", hue="Sex", size=3)
g.map(sns.kdeplot, "Age", shade=True)
sns.despine(top=True, right=True, left=True,bottom=True)
g.add_legend()
plt.show()
```
